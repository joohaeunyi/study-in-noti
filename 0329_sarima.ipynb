SARIMA

1. SARIMA의 정의
   - SARIMA는 Seasonal AutoRegressive Integrated Moving Average의 약자입니다.
   - 시계열 데이터의 계절성을 고려하여 예측을 수행하는 통계적 모델입니다.
   - ARIMA 모델의 확장된 버전으로, 계절성을 추가로 다룰 수 있습니다.


2. SARIMA 모델의 구성요소
   - SARIMA 모델은 세 가지 주요 구성요소를 가지고 있습니다:
         a. Seasonal (계절성): 데이터의 계절적 패턴을 나타냅니다.
         b. AutoRegressive (자기회귀): 과거 관측값이 현재 값에 영향을 미치는 것을 모델링합니다.
         c. Inter (차분): 데이터의 불안정성을 제거하기 위해 차분을 수행합니다.
         d. Moving Average (이동평균): 과거 예측 오차가 현재 값에 영향을 미치는 것을 모델링합니다.
   - 각 구성요소는 계절성과 비계절성 부분으로 나뉘어 표현됩니다.
   

3. SARIMA 모델의 표기법
   - SARIMA 모델은 (p, d, q)(P, D, Q)m으로 표기됩니다.
         a. p: 비계절성 자기회귀 차수
         b. d: 비계절성 차분 차수
         c. q: 비계절성 이동평균 차수
         d. P: 계절성 자기회귀 차수
         e. D: 계절성 차분 차수
         f. Q: 계절성 이동평균 차수
         g. m: 계절성의 주기 (예: 월별 데이터의 경우 m=12)
     

4. SARIMA 모델의 장점
   - 계절성을 고려하여 예측할 수 있어 정확도가 높습니다.
   - 단기 및 장기 예측 모두에 적합합니다.
   - 다양한 시계열 데이터에 적용 가능합니다.
   - 모델의 해석력이 좋아 결과를 이해하기 쉽습니다.
   

5. SARIMA 모델의 적용 분야
   - 경제 및 금융 예측 (주가, 환율, 판매량 등)
   - 수요 예측 (전력 수요, 제품 수요 등)
   - 기상 및 환경 예측 (온도, 습도, 대기오염 등)
   - 의료 및 보건 분야 (질병 발생률, 입원 환자 수 등)
